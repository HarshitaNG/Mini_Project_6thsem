The below reads the gsheet and stores it as a dataframe.

```python
gc = gspread.authorize(creds)
worksheet = gc.open('Copy of Blank Quiz (Responses).xlsx').sheet1

# get_all_values gives a list of rows.
rows = worksheet.get_all_values()

# Use the first row as the column headers.
headers = rows[0]
data = pd.DataFrame(rows[1:], columns=headers)
```
